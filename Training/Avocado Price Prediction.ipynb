# Splitting

```python
x = df.drop('AveragePrice',axis=1)
y=df['AveragePrice']
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state = 42)
```

# Standardization

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
x_train = ss.fit_transform(x_train)
x_test = ss.fit_transform(x_test)
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
from sklearn.metrics import r2_score
LR = LinearRegression()
LR.fit(x_train, y_train)

y_pred_test_linear = LR.predict(x_test)
linear_r2_test = r2_score(y_test, y_pred_test_linear)
print("Linear Regression Test R-squared:", linear_r2_test)
training_r2_linear = LR.score(x_train, y_train)
print("Linear Regression Training R-squared:", training_r2_linear)
```

```python
cv_scores_linear_train = cross_val_score(LR, x_train, y_train, cv=5, scoring='r2')

print("Cross-Validation R-squared scores on Training set:", cv_scores_linear_train)
print("Cross-Validation R-squared on Training set:", np.mean(cv_scores_linear_train))
```

```python
cv_scores_linear_test = cross_val_score(LR, x_test, y_test, cv=5, scoring='r2')

print("Cross-Validation R-squared scores on Test set:", cv_scores_linear_test)
print("Cross-Validation R-squared on Test set:", np.mean(cv_scores_linear_test))
```

# Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
DTR = DecisionTreeRegressor(criterion='poisson', max_depth=5, random_state=7)
DTR.fit(x_train, y_train)
y_pred_test_dtr = DTR.predict(x_test)

dtr_r2_test = r2_score(y_test, y_pred_test_dtr)
print("Decision Tree Test R-squared:", dtr_r2_test)
training_r2_dtr = DTR.score(x_train, y_train)
print("Decision Tree Training R-squared:", training_r2_dtr)
```

```python
cv_scores_dtr_train = cross_val_score(DTR, x_train, y_train, cv=5, scoring='r2')

print("Cross-Validation R-squared scores on Training set:", cv_scores_dtr_train)
print("Cross-Validation R-squared on Training set:", np.mean(cv_scores_dtr_train))
```

```python
cv_scores_dtr_test = cross_val_score(DTR, x_test, y_test, cv=5, scoring='r2')

print("Cross-Validation R-squared scores on Test set:", cv_scores_dtr_test)
print("Cross-Validation R-squared on Test set:", np.mean(cv_scores_dtr_test))
```

# Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
RF = RandomForestRegressor(n_estimators=100, max_depth=5, random_state=7)
RF.fit(x_train, y_train)
y_pred_test_rf = RF.predict(x_test)

rf_r2_test = r2_score(y_test, y_pred_test_rf)
print("Random Forest Test R-squared:", rf_r2_test)
training_r2_rf = RF.score(x_train, y_train)
print("Random Forest Training R-squared:", training_r2_rf)
```

```python
cv_scores_rf_train = cross_val_score(RF, x_train, y_train, cv=5, scoring='r2')

print("\nRandom Forest Cross-Validation R-squared scores on Training set:", cv_scores_rf_train)
print("Average Random Forest Cross-Validation R-squared on Training set:", np.mean(cv_scores_rf_train))
```

```python
cv_scores_rf_test = cross_val_score(RF, x_test, y_test, cv=5, scoring='r2')

print("\nRandom Forest Cross-Validation R-squared scores on Test set:", cv_scores_rf_test)
print("Average Random Forest Cross-Validation R-squared on Test set:", np.mean(cv_scores_rf_test))
```

# XGBoost 

```python
import xgboost as xgb
regressor = xgb.XGBRegressor(n_estimators=1000, learning_rate=0.1, max_depth=3, objective='reg:squarederror', random_state=7)
regressor.fit(x_train, y_train)
y_pred_test = regressor.predict(x_test)

xgb_r2_test = r2_score(y_test, y_pred_test)
print("XGBoost Test R-squared:", xgb_r2_test)
training_r2 = regressor.score(x_train, y_train)
print("XGBoost Training R-squared:", training_r2)
```

```python
cv_scores_xgb_train = cross_val_score(regressor, x_train, y_train, cv=3, scoring='r2')

print("XGBoost Cross-Validation R-squared scores on Training set:", cv_scores_xgb_train)
print("Average XGBoost Cross-Validation R-squared on Training set:", np.mean(cv_scores_xgb_train))
```

```python
cv_scores_xgb_test = cross_val_score(regressor, x_test, y_test, cv=3, scoring='r2')

print("\nXGBoost Cross-Validation R-squared scores on Test set:", cv_scores_xgb_test)
print("Average XGBoost Cross-Validation R-squared on Test set:", np.mean(cv_scores_xgb_test))
```

# KNN

```python
from sklearn.neighbors import KNeighborsRegressor
KNN = KNeighborsRegressor(n_neighbors=15)
KNN.fit(x_train, y_train)
y_pred_test_knn = KNN.predict(x_test)

knn_r2_test = r2_score(y_test, y_pred_test_knn)
print("KNN Test R-squared:", knn_r2_test)
training_r2_knn = KNN.score(x_train, y_train)
print("KNN Training R-squared:", training_r2_knn)
```
